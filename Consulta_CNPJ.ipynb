```python
import pandas as pd
import requests
import time

cnpj_df = pd.DataFrame({'cnpj': ['16811931000100', '60872504000123', '16670085000155', '33014556000196', '47960950000121']})
cnpj_data_list = []

def get_cnpj_info(cnpj):
    url = f"https://publica.cnpj.ws/cnpj/{cnpj}"
    resp = requests.get(url)

    if resp.status_code == 200:
        data = resp.json()
        df = pd.json_normalize(data)
        return df
    elif resp.status_code == 429:
        print(f"Limite excedido. Espere por 60 segundos.")
        time.sleep(60)
        return get_cnpj_info(cnpj)
    else:
        print(f"Error para o CNPJ {cnpj}. Status Code: {resp.status_code}")
        print(f"Error content: {resp.text}")
        return None

for cnpj in cnpj_df['cnpj']:
    cnpj_data = get_cnpj_info(cnpj)

    if cnpj_data is not None:
        cnpj_data_list.append(cnpj_data)

final_cnpj_df = pd.concat(cnpj_data_list, ignore_index=True)
final_cnpj_df.to_excel('listacnpj.xlsx', index=False)
final_cnpj_df
```
